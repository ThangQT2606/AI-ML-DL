2. Tiền xử lý dữ liệu

Mã hóa dữ liệu

```python
df2 = df.copy()
```

```python
import category_encoders as ce
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
df2["Credit_Score"] = encode.fit_transform(df2["Credit_Score"])
Credit_Labels = encode.classes_
print(Credit_Labels)
```

```python
object_columns = df2.select_dtypes('object').columns.to_list()
categori_columns = ['Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Occupation']
one_encode = ce.OneHotEncoder(cols=categori_columns, return_df=True)
df2 = one_encode.fit_transform(df2)

for remo in categori_columns:
    if remo in object_columns:
        object_columns.remove(remo)

encode = LabelEncoder()
for obj in object_columns:
    df2[obj] = encode.fit_transform(df2[obj])
```

```python
object_columns
```

Loại bỏ feature có tính tương quan cao

```python
correlation_matrix = df2.corr().abs()

# Tạo một mas với các giá trị True ở đường chéo trên ma trận tương quan
upper_tri = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(np.bool_))

# Tìm các cặp đặc trưng có tương quan cao
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.8)]
print(to_drop)
# Loại bỏ các đặc trưng có tương quan cao từ dữ liệu
df2 = df2.drop(to_drop, axis=1)
```

```python
df2
```

Xử lý ngoại lai (Nhiễu)

```python
process_noise = ["Num_Credit_Inquiries", "Delay_from_due_date", "Outstanding_Debt",
                "Monthly_Balance", "Annual_Income", "Changed_Credit_Limit", "Credit_Utilization_Ratio", "Total_EMI_per_month"]
for _encode in process_noise:
    q25, q75 = np.quantile(df2[_encode], 0.25), np.quantile(df2[_encode], 0.75)
    iqr = q75 - q25
    lower, upper = q25 - 0.75*iqr, q75 + 0.75*iqr
    df2 = df2[(df2[_encode] < upper) & (df2[_encode] > lower)]
```

Tách bộ dữ liệu thành Feature và Nhãn

```python
x = df2.drop("Credit_Score", axis=1)
y = df2["Credit_Score"]
features = x.columns.to_list()
```

```python
x.shape, y.shape
```

Thêm dữ liệu sau khi xử lý ngoại lai

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
smote = RandomOverSampler(random_state=42)
x, y = smote.fit_resample(x, y)
```

Chuẩn hóa dữ liệu

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# sc = MinMaxScaler(feature_range=(-1, 1))
sc = StandardScaler()
sc.fit(x)
x = sc.transform(x)
```

```python
x.shape, y.shape
```

Giảm chiều dữ liệu (Nếu cần thiết)

```python
from sklearn.decomposition import PCA
n_comp = 30
pca = PCA(n_components=n_comp)
x_pca = pca.fit_transform(x)
```

```python
counter = 0
printed_features = set()
print("Các feature còn lại sau khi giảm chiều:")
for i, component in enumerate(pca.components_):  
    for j, feature_index in enumerate(np.argsort(np.abs(component))[::-1][:n_comp]):
        feature_name = features[feature_index]
        if feature_name not in printed_features and len(printed_features) < n_comp:
            printed_features.add(feature_name)
for feature_i in printed_features:
    counter += 1
    print(f"{feature_i}", end= ", ")  
    if counter % 5 == 0:
        print()
print(len(printed_features))
```

Chia bộ dữ liệu thành 2 tập train và test với tỉ lệ 80 : 20

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42, shuffle= True, stratify= y)
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

3. Xây dựng mô hình RandomForest cho bài toán phân loại

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state= 42, class_weight='balanced')
model.fit(x_train, y_train)
```

4. Đánh giá mô hình

```python
_score = model.score(x_test, y_test)
print(_score)
```

```python
y_pred = model.predict(x_test)
```
